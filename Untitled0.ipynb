```python
with open('sa11.txt', 'r') as f:
  for line in f.readlines():
    time, iowait  = (line.split()[0], float(line.split()[6]))
    if iowait > 43:
      print(f'{time} {iowait}')
```

```python
time_list = []
spent_time = []
with open('dmesg1', 'r') as f:
  for line in f.readlines():
    time_spent_for_line = line.split()[1].split(']')[0]
    time_list.append(float(time_spent_for_line))
    #print(time_spent_for_line)
  
  for i in range(len(time_list)):
    if i == 0:
      continue
    spent_time.append(round((time_list[i] - time_list[i-1])*1000,4))
print(sorted(spent_time))

```
